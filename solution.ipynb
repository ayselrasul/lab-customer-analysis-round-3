```python
%matplotlib notebook

#Show a plot of the total number of responses

data=pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')

data1 = data[['Response']].value_counts()
data1.plot(kind='bar')
```

```python
%matplotlib notebook

#Show a plot of the response rate by the sales channel

data=pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')
data=data.groupby(['Response','Sales Channel']).size()
data=data.unstack()
data.plot(kind='bar')
```

```python
%matplotlib notebook

#Show a plot of the response rate by income

data=pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')

data3=data.groupby(['Response','Income']).size()
data3 = data3.reset_index()


ax = data3[['Response','Income']].plot.box()

```
